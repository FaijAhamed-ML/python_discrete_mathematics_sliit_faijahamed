```python
A = {'U1', 'U2', 'U3', 'U4', 'U5', 'U6'}
B = {'U3', 'U4', 'U5', 'U7'}
C = {'U5', 'U7'}
```

```python
is_subset = C.issubset(A)
print(f"Is C a subset of A? {is_subset}")

is_proper_subset = C < A 
print(f"Is C a proper subset of A? {is_proper_subset}")
```

```python
both_access = A.intersection(B) 
print(f"Elements: {both_access}")
```

```python
file_not_admin = B.difference(C) 
print(f"Elements: {file_not_admin}")
```

```python
are_disjoint = A.isdisjoint(C)
print(f"Are A and C disjoint? {are_disjoint}")
```

```python
at_least_one = A.union(B) 
print(f"Elements: {at_least_one}")
```

```python
exclusive_access = A.symmetric_difference(B)
print(f"Elements: {exclusive_access}")
```

```python
scores = [78, 85, 95, 62, 88, 75, 95, 85, 62, 90]
print(f"Original scores: {scores}")
```

```python
unique_scores = list(set(scores))
print(f"Unique scores: {unique_scores}")
```

```python
lowest_score = min(scores)
scores.remove(lowest_score) 
scores.append(99)
print(f"Updated list (lowest removed, 99 added): {scores}")
```

```python
unique_scores.sort(reverse=True)

top_3_highest = unique_scores[:3]
print(f"Top 3 highest unique scores (descending): {top_3_highest}")
```

```python
products = {
    101: ["Laptop", "Electronics", 15, 55000],
    102: ["Table", "Furniture", 30, 4500],
    103: ["Pen", "Stationery", 100, 20],
    104: ["Mobile Phone", "Electronics", 25, 15000],
    105: ["Chair", "Furniture", 50, 2500],
    106: ["Notebook", "Stationery", 200, 50],
    107: ["Headphones", "Electronics", 40, 2500],
    108: ["Water Bottle", "Kitchenware", 60, 350]
}
```

```python
for details in products.values():
    if details[1] == "Electronics":
        print(details[0])
```

```python
for details in products.values():
    if details[1] == "Furniture" and details[3] > 3000:
        print(details[0])
```

```python
fever = {201, 202, 203, 204, 205, 206, 207}
cough = {203, 204, 205, 208, 209, 210}
fatigue = {202, 204, 206, 209, 211, 212}
```

```python
all_three = fever.intersection(cough, fatigue)
print(f"Patients: {all_three}")
```

```python
fever_only = fever.difference(cough.union(fatigue))
print(f"Patients: {fever_only}")
```

```python
fever_and_cough = fever.intersection(cough)
fever_and_fatigue = fever.intersection(fatigue)
cough_and_fatigue = cough.intersection(fatigue)

at_least_two = fever_and_cough.union(fever_and_fatigue, cough_and_fatigue)
count = len(at_least_two)
print(f"Number of patients: {count} (IDs: {at_least_two})")
```

```python
prioritized = fever.intersection(cough.union(fatigue))
print(f"Patients: {prioritized}")
```

```python
import itertools
```

```python
F = ['React', 'Angular']
B = ['Node.js', 'Django']
D = ['Docker', 'Nginx']
```

```python
tech_stacks = list(itertools.product(F, B))

for stack in tech_stacks:
    print(stack)
```

```python
deployment_subsets = []

for i in range(len(D) + 1):
    combo = itertools.combinations(D, i)
    deployment_subsets.extend(list(combo))

for subset in deployment_subsets:
    if not subset:
        print("<-- Empty set -->")
    else:
        print(subset)
```
